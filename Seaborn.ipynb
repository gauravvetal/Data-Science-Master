```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
'''
Scatter Plot: Visualizes the relationship between two continuous variables.
Histogram: Represents the distribution of a single continuous variable.
Bar Plot: Shows the relationship between a categorical variable and a continuous variable.
Box Plot: Displays the distribution of a continuous variable, especially when comparing categories.
Heatmap: Highlights relationships between two categorical or continuous variables, often used for correlation matrices or pattern identification.
'''
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event', style='region')
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
titanic=sns.load_dataset('titanic') 
titanic
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x = 'pclass', y = 'age', data = titanic)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
```

```python
sns.boxplot(x = 'pclass', y = 'fare', data = titanic)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
diamonds=sns.load_dataset('diamonds')
diamonds
```

```python
sns.displot(data=diamonds,x='price',hue='cut',kind='hist')
```

```python
#Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
iris=sns.load_dataset('iris')
iris
```

```python
sns.set(style="ticks")
sns.pairplot(data=iris, hue="species", markers=["o", "s", "D"])
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
flight=sns.load_dataset('flights')
flight
```

```python
flights_pivot = flight.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
```
