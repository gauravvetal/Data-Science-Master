```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
'''
Scatter plots are used to represent individual data points as markers on a 2D plane. 
They are useful for visualizing the relationships between two variables and identifying patterns or clusters.
'''
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
'''
subplot in Matplotlib is used to create and arrange multiple subplots (smaller plots) within a single figure.
'''
x = np.array([0, 1, 2, 3, 4, 5])
plt.subplot(2,2,1)
plt.plot(x,np.array([0, 100, 200, 300, 400, 500]))
plt.title('Line 1')

plt.subplot(2,2,2)
plt.plot(x,np.array([50, 20, 40, 20, 60, 70]))
plt.title('Line 2')

plt.subplot(2,2,3)
plt.plot(x,np.array([10, 20, 30, 40, 50, 60]))
plt.title('Line 3')

plt.subplot(2,2,4)
plt.plot(x,np.array([200, 350, 250, 550, 450, 150]))
plt.title('Line 4')

plt.tight_layout()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
'''
Bar charts (or bar plots) are used to represent categorical data with rectangular bars. 
They are commonly used to compare and display data for different categories or groups.
'''
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.subplot(1,2,1)
plt.bar(company,profit)

plt.subplot(1,2,2)
plt.barh(company,profit)
plt.tight_layout()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
'''

A box plot in Matplotlib is a graphical representation that displays the distribution, central tendency, and spread of a dataset, typically with a box-and-whisker diagram, and it can be created using the plt.boxplot() function.
'''
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
```
